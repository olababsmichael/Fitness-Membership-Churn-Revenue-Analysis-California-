```python
# Convert check-in times to time objects
df["avg_time_check_in"] = pd.to_datetime(df["avg_time_check_in"],  format="%H:%M:%S").dt.time

# Convert check-out times to time objects
df["avg_time_check_out"] = pd.to_datetime(df["avg_time_check_out"],  format="%H:%M:%S").dt.time
```

```python
# Convert a Series to nullable boolean values (True/False/NA) based on common text representations
def to_nullable_boolean(series):
    s = series.copy()
    s = s.astype(str).str.strip().str.lower().str.replace({
        "true": True, "false": False,
        "yes": True, "no": False,
        "1": True, "2": False,
        "nan": pd.NA, "none": pd.NA, "": pd.NA
    })
```

FEATURE ENGINEERING

```python
# membership_tenure_months = months between join_date and today.

df["join_date"] = pd.to_datetime(df["join_date"])
today = pd.to_datetime(datetime.today().date())  
df["membership_tenure_months"] = (
    (today.year - df["join_date"].dt.year) * 12 +
    (today.month - df["join_date"].dt.month)
)
df["membership_tenure_months"]
```

```python
# Ensure last_visit_date is in datetime format
df["last_visit_date"] = pd.to_datetime(df["last_visit_date"], errors="coerce")

# Get today's date
today = pd.to_datetime(datetime.today().date())

# Calculate days since last visit (handle missing values)
df["days_since_last_visit"] = (today - df["last_visit_date"]).dt.days
df["days_since_last_visit"]

```

```python
 df["total_visits_per_month"] = df["visit_per_week"] * 4
df["total_visits_per_month"] 
```

```python
df["average_visit_duration_hours"] = (df["duration_in_gym_minutes"] / 60).round(1)
df["average_visit_duration_hours"] 
```

```python
df["discount_amount"] = (df["adjusted_price"] - df["final_price"]).round(2)
df["discount_amount"]
```

```python
df["is_churned"] = (df["days_since_last_visit"] > 60).astype(int)
df["is_churned"]
```
